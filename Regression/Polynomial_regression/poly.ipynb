# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

# Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
dataset.head()
```

```python
X = dataset.iloc[:,1].values
y = dataset.iloc[:,-1].values
X = X.reshape(-1,1)
y = y.reshape(-1,1)
```

```python

X
```

```python
y
```

# Training the Linear Regression model on the whole dataset

```python
line_regress = LinearRegression()
line_regress.fit(X,y)
```

# Training the Polynomial Regression model

```python
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
X_poly
```

```python
line_regress2 = LinearRegression()
line_regress2.fit(X_poly,y)
```

# Visuliasing the Linear Regression results

```python
plt.scatter(X,y, color='red', label='Actual data')
plt.plot(X, line_regress.predict(X),color='blue', label='Predicted result')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Predicting salaries of employee based on position within company')
plt.legend()
plt.show()
```

# Visuliasing the Polynomial Regression results

```python
plt.scatter(X,y, color='red', label='Actual data')
plt.plot(X, line_regress2.predict(X_poly),color='blue', label='Predicted result')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Predicting salaries of employee based on position within company')
plt.legend()
plt.show()
```

# Predicting a new result with Linear Regression

```python
print(line_regress.predict([[6.5]]))
```

# Predicting a new result with Polynomial Regression

```python
print(line_regress2.predict(poly_reg.fit_transform([[6.5]])))
```
