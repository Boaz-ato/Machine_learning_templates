# Importing the libaries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.ensemble import RandomForestRegressor
```

# Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1].values.reshape(-1,1)
y = dataset.iloc[:,2].values.reshape(-1,1)
```

```python
X
```

```python
y
```

# Training the Random Forest Regression model on the whole dataset

```python
regr = RandomForestRegressor(n_estimators=10, random_state=0)
regr.fit(X,y)
```

# Predicting a new value

```python
regr.predict([[6.5]])
```

# Visualising the Random Forest Regreesion results (higher resolution)

```python
X_pred = np.arange(start=0, stop=10, step=0.01)
y_pred = regr.predict(X_pred.reshape(-1,1))
plt.scatter(X, regr.predict(X), label='True value')
plt.plot(X_pred, y_pred, label='Predicted value')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.title('Position vs Salary(Random Forest regression)')
plt.legend()
plt.show()
```
