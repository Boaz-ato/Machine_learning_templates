# Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.tree import DecisionTreeRegressor
```

# Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1].values.reshape(-1,1)
y = dataset.iloc[:,-1].values.reshape(-1,1)
```

```python
X
```

```python
y
```

# Training the Decision Tree Regression model on the whole dataset

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X,y)
```

# Predicting a new result

```python
regressor.predict([[6.5]])
```

# Visualising the Decision Tree Regression results (higher resolution)

```python
X_pred = np.linspace(start=0, stop= 10,num= 100)
y_pred = regressor.predict(X_pred.reshape(-1,1))
plt.scatter(X, regressor.predict(X), label='True values')
plt.plot(X_pred, y_pred, label='Predicted values ')
plt.title('Using Decision Tree to predict the salaries of position in a company')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
